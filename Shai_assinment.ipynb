***Part 1 Basic Data Exploration:***

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Salaries.csv', error_bad_lines=False)
df.head()
```

```python
# Identify the number of rows and columns
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Determine the data types of each column
data_types = df.dtypes
print("\nData types of each column:")
print(data_types)
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

**Part2 Descriptive Statistics:**


```python
# Calculate mean
mean_TotalPay = df['TotalPay'].mean()
print(f"Mean TotalPay: {mean_TotalPay}")
```

```python
# Calculate median
median_TotalPay = df['TotalPay'].median()
print(f"Median TotalPay: {median_TotalPay}")
```

```python
# Calculate mode
mode_TotalPay = df['TotalPay'].mode()
print(f"Mode TotalPay: {mode_TotalPay}")
```

```python
# Calculate minimum and maximum TotalPay

min_TotalPay  = df['TotalPay'].min()
max_TotalPay  = df['TotalPay'].max()
print(f"Minimum TotalPay : {min_TotalPay }")
print(f"Maximum TotalPay : {max_TotalPay }")
```

```python
# Determine the range of TotalPay
TotalPay_range = max_TotalPay - min_TotalPay
print(f"TotalPay Range: {TotalPay_range}")
```

```python
# Calculate standard deviation
std_dev_TotalPay = df['TotalPay'].std()
print(f"Standard Deviation of TotalPay: {std_dev_TotalPay}")
```

```python
## Additional summary statistics using describe() also we can use it instead the previous
summary_stats = df['TotalPay'].describe()
print("\nSummary Statistics:")
print(summary_stats)
```

***part3 Data Cleaning:***

```python
# hence Benefits = TotalPayBenefits - TotalPay
df['Benefits'] = df['TotalPayBenefits']-df['TotalPay']
df.head(20)
```

```python
#we used missing indicator method to create a binary indicator variable to denote whether a value was missing
#for columns Notes and status beacause the datatpe of them are float64.

df['Notes'] = df['Notes'].isnull().astype(int)
df['Status'] = df['Status'].isnull().astype(int)
df.head(20)

```

**Part4 Basic Data Visualization**

```python
# Histogram for TotalPay Distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('TotalPay Distribution')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram for TotalPayBenefits Distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPayBenefits'], bins=20, color='skyblue', edgecolor='black')
plt.title('TotalPayBenefits Distribution')
plt.xlabel('TotalPayBenefits')
plt.ylabel('Frequency')
plt.grid(True)
plt.show();
```

```python
# Bar Chart for Proportion of Employees according to JobTitle
plt.figure(figsize=(10, 8))
JobTitle_counts = df['JobTitle'].value_counts()
plt.bar(JobTitle_counts.index,JobTitle_counts.values, color='lightcoral')
plt.title('Proportion of Employees Employees according to JobTitle')
plt.xlabel('JobTitle')
plt.ylabel('Number of Employees')
plt.grid(axis='y')
plt.show()
```

```python
# Pie Chart for Proportion of Employees in Different Departments
plt.figure(figsize=(10, 8))
plt.pie(JobTitle_counts, labels=JobTitle_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

**Part5Grouped Analysis**

```python
# Group by the 'OvertimePay' column and calculate the average TotalPay for each group
grouped_data = df.groupby('OvertimePay')['TotalPay'].mean()
# Display the grouped data with average TotalPay
print("Average TotalPay  by OvertimePay:")
print(grouped_data)
# Group by 'OvertimePay' and calculate multiple summary statistics
grouped_data = df.groupby('OvertimePay')['TotalPay'].agg(['mean', 'median', 'std', 'count'])

# Display the grouped data with various statistics
print("Summary Statistics by OtherPay:")
print(grouped_data)
```

```python
# Group by the 'OtherPay' column and calculate the average TotalPay for each group
grouped_data = df.groupby('OtherPay')['TotalPay'].mean()
# Display the grouped data with average TotalPay
print("Average TotalPay  by OtherPay:")
print(grouped_data)
# Group by 'OtherPay' and calculate multiple summary statistics
grouped_data = df.groupby('OtherPay')['TotalPay'].agg(['mean', 'median', 'std', 'count'])

# Display the grouped data with various statistics
print("Summary Statistics by OtherPay:")
print(grouped_data)
```

**Part6 Simple Correlation Analysis**

```python
# Calculate the correlation coefficient
correlation_coefficient = df['TotalPay'].corr(df['BasePay'])

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['BasePay'], df['TotalPay'], color='skyblue')
plt.title(f'Scatter Plot: TotalPay vs BasePay (Correlation: {correlation_coefficient:.2f})')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
